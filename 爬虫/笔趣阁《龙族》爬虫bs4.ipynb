```python
import requests
from bs4 import BeautifulSoup
#需求，爬取笔趣阁《龙族》所有章节目录以及详细内容
```

```python
url = "https://www.jx.la/book/87705/"
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.121 Safari/537.36'
    }
res = requests.get(url = url, headers = headers)
print(res.encoding)
res = res.text
```

```python
soup = BeautifulSoup(res,'lxml')
#解析章节标题和详情页的url
li_list = soup.select('.section-box >ul > li')
for li in li_list:
    print(li.a.text)
```

```python
https://www.jx.la/book/87705/4652657.html
```

```python
fp = open('./longzu.txt','w',encoding='utf-8')
for li in li_list:
    title = li.a.text
    new_url = 'https://www.jx.la'+li.a['href']
    new_res = requests.get(url = new_url,headers = headers).text
    
    #解析出章节页的详细信息
    new_soup = BeautifulSoup(new_res,'lxml')
    div_cont = new_soup.find('div',class_ = 'content')
    content = div_cont.text
    print(content)
    fp.write(title+':'+content+'\n')
    print(title,'爬取成功！！！')
    
```
